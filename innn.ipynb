```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
from datetime import datetime 
sns.set()
INvideos = pd.read_csv('INvideos.csv')
```

```python
df_INvideos = pd.read_csv('INvideos.csv')
df_INvideos[['video_id','trending_date','title','channel_title','category_id','publish_time','tags','views','likes','dislikes','comment_count','thumbnail_link','comments_disabled','ratings_disabled','video_error_or_removed','description']]
```

```python
INvideos.drop( 'video_id',axis =1,inplace =True)
## the video id column is dropped becouse it does not present enough value when it comes to data analysis
```

```python
INvideos.drop( 'thumbnail_link',axis =1,inplace =True)
## we drop the thumnail link as well becouse it does not present enough value when it comes to data analysis
```

```python
#The following visualization identifies the total views per day between 11/2017-01/2018.
ytDf_ttrendDate = df_INvideos["trending_date"]
ytDf_views = df_INvideos["views"]

order = np.argsort(ytDf_ttrendDate )
x_trendDate = np.array(ytDf_ttrendDate)[order]
y_views = np.array(ytDf_views )[order]
```

```python
print("During the period of October 2017 until January 2018, \
      we had a total of %d views. In the following distribution, \
      we can see the total views per day" %  (df_INvideos["views"].sum()))

plt.figure(1)
plt.subplot(211)


plt.title("Total Views Per day (11/2017 to 1/2018)")
plt.xticks(rotation = "vertical")
plt.xlabel("Time")
plt.ylabel("Views")
plt.plot(x_trendDate,y_views)
plt.show()
#We observe a major increase of Youtube views from beginning of December 2017 until its end.
```

```python
INvideos['likes_log'] = np.log(INvideos['likes']+1)
INvideos['views_log'] = np.log(INvideos['views'] +1)
INvideos['dislikes_log'] = np.log(INvideos['dislikes'] +1)
INvideos['comment_count_log'] = np.log(INvideos['comment_count']+1)
```

```python
plt.figure(figsize = (12,6))
plt.subplot(221)
g1 = sns.distplot(INvideos['likes_log'], color = 'green')
g1.set_title("LIKES LOG DISTRIBUTION", fontsize = 16)

plt.subplot(222)
g2 = sns.distplot(INvideos['views_log'])
g2.set_title("VIEWS LOG DISTRIBUTION", fontsize = 16)

plt.subplot(223)
g3 = sns.distplot(INvideos['dislikes_log'], color = 'r')
g3.set_title("DISLIKES LOG DISTRIBUTION", fontsize=16)

plt.subplot(224)
g4 = sns.distplot(INvideos['comment_count_log'])
g4.set_title("COMMENT COUNT LOG DISTRIBUTION", fontsize=16)

plt.subplots_adjust(wspace = 0.2, hspace = 0.4, top = 0.9)

plt.show()
#the graph below shows 4 charts for the distribution for 1.likes_log , 2.dislikes_log , 3.views_log , 4.comments_count_log
#the average was between 1. 10 for likes_log , 2. 6.5 for dislike_log , 3. 14 for views_log , 4. 7 for comments_count_log
```
